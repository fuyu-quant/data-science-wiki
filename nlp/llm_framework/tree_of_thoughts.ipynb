### Tree of Thoughts

```python
args = argparse.Namespace(backend='gpt-4', 
                          temperature=0.7,
                          n_generate_sample=1,
                          n_evaluate_sample=3,
                          n_select_sample=5,
                          method_generate='propose',
                          method_evaluate='value',
                          method_select='greedy'
                          )
```

```python
task = Game24Task()

# Tree of Thoughts
ys, infos = solve(args, task, 900)
```
